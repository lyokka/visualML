```python
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import pandas as pd
import numpy as np


init_notebook_mode(connected=True)
```

```python
Data = pd.read_csv('train.csv')
X = Data[['LotArea','LotFrontage']]
Y = Data['SalePrice']

#plot in 3d
trace1 = go.Scatter3d(
    x=X['LotArea'],
    y=X['LotFrontage'],
    z=Y,
    mode='markers',
    marker=dict(
        size=1,
        line=dict(
            color='rgba(217, 217, 217, 0.14)',
            width=0.5
        )
    )
)

layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    ),
    scene=go.Scene(
        xaxis=go.XAxis(title='LotArea'),
        yaxis=go.YAxis(title='LotFrontage'),
        zaxis=go.ZAxis(title='SalePrice')
    )
)
fig = go.Figure(data=[trace1], layout=layout)
iplot(fig, filename='LotArea, LotFrontage with SalePrice')
```

```python
X = Data['LotArea']

#Create a trace
trace = go.Scatter(
    x = X,
    y = Y,
    mode = 'markers',
    marker = dict(
        size = 3,
        line = dict(
            width = 1,
            color = 'rgb(0, 0, 0)'
        )
    )
)

layout = go.Layout(
    xaxis= dict(
        title= 'LotArea',
        ticklen= 5,
        zeroline= False,
        gridwidth= 2,
    ),
    yaxis=dict(
        title= 'SalePrice',
        ticklen= 5,
        gridwidth= 2,
    )
)
data = [trace]

# Plot and embed in ipython notebook!
fig = go.Figure(data=[trace], layout=layout)
iplot(fig, filename='LotArea v.s SalePrice')
```
