
# LMS least mean square

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

data = pd.read_csv('train.csv')
X = data[['LotArea','LotFrontage']]
Y = data['SalePrice']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X['LotArea'], X['LotFrontage'], Y, c = 'blue', marker = 'o')
ax.set_xlabel('LotArea')
ax.set_ylabel('LotFrontage')
ax.set_zlabel('SalePrice')
plt.show()
```

```python
X = data['LotArea']
plt.scatter(X, Y, c = 'blue')
plt.xlabel('LotArea')
plt.ylabel('SalePrice')
plt.show()
```

```python
#analytic solution
from numpy.linalg import inv
X = data['LotArea']
plt.scatter(X, Y, c = 'blue')
N = len(X)
#initialization
Theta = np.array([1000, 1000])
Y_hat = np.ones(N)
alpha = 1e-12
ones = pd.DataFrame(np.ones([N, 1]))
X = ones.join(X)

ThetaOpt = np.dot(np.dot(inv(np.dot(np.transpose(X),X)), np.transpose(X)), Y)
x = pd.DataFrame(np.arange(-5000, 250000, 100), columns = ['0'])
N = len(x)
ones = pd.DataFrame(np.ones([N, 1]), columns = ['1'])
x = ones.join(x)
y_hat = np.dot(x, ThetaOpt)
plt.plot(x['0'], y_hat, c = 'red')
plt.show()
```

```python
#loss function
def J(Y_hat, Y):
    return np.dot((Y_hat - Y), np.transpose(Y_hat - Y))/2
Y_hat = np.dot(X, ThetaOpt)
print(J(Y_hat, Y))
print(ThetaOpt)
```

```python
#initialization
Theta = np.array([1000, 1000])
Y_hat = np.ones(N)
alpha = 6.5e-12
ones = pd.DataFrame(np.ones([N, 1]))
X = ones.join(X)

while J(Y_hat, np.asarray(Y)) > 4.2834599e+12:
    Y_hat = np.dot(X, Theta)
    Theta = Theta + alpha*np.dot((np.asarray(Y) - Y_hat), X)
    print(J(Y_hat, np.asarray(Y)))
```
