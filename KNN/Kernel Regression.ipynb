# Kernel regression
Kernel: weighting function is called a kernel and it measures the similaity between examples 

gaussian kernel: $K(x,x_i) = exp (\frac{-d^{2}(x,x_i)}{\sigma^{2}})$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('data1.csv', header = None)
X = data[0]
Y = data[1]

x = np.linspace(-35, 35, 1000)

# dist function
def euclideanDist(a, b):
    return abs(a - b)

sigma = 0.1
y_hat = []
for i in x:
    K = []
    for j in X:
        K.append(np.exp(-euclideanDist(i, j)**2 / sigma))
    y_hat.append(sum(K*Y)/sum(K))

plt.scatter(X, Y)        
plt.plot(x, y_hat, 'red')
plt.show()
```
