```python
# http://learning.cis.upenn.edu/cis520_fall2009/index.php?n=Lectures.LocalLearning
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('data1.csv', header = None)

X = np.asarray(data[0])
Y = np.asarray(data[1])

x = np.linspace(-35, 35, 1000)
k = 20 # k nearest neighbors

def euclideanDist(a, b):
    return abs(a - b)
y_hat = []
for i in x:
    edist = []
    for j in X:
        edist.append(euclideanDist(i, j))
    # sort the distance
    k_ind = sorted(range(len(edist)), key=lambda m: edist[m])
    # average the k nearest neighbors 
    y_hat.append(sum(Y[k_ind][:k])/k)#shortcoming of KNN, equally weighted
    
plt.scatter(data[0], data[1])
plt.plot(x, y_hat, c = 'red')
plt.show()
```
